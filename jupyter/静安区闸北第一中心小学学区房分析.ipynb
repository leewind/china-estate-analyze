```python
import pandas as pd
```

# 读取房产数据并且对空数据做处理

```python
# 2019.12.5数据
df = pd.read_csv('../data/daily/china_estate_get_one_day_20191205.csv', names=[
    'city', 'district', 'hotpot', 'link', 'desc', 'odesc', 'level', 'build_year','structure', 'size', 'direction', 'price', 'tax', 'date', 'name'
])
df = df[df['city']=='上海']
df = df.drop(['link', 'odesc'], 1).reset_index().drop(['index'], 1)
df.head(10)
```

```python
df['tax'] = df['tax'].fillna('未知')
```

```python
df['price_per_m2'] = df['price'] / df['size']
```

# 根据学区规划获取片区

根据2019年3月[上海教育](http://edu.sh.gov.cn/html/article/201903/99860.html)提供的学区划片数据显示，上海市静安区闸北第一中心小学的片区如下：

> 上海市静安区闸北第一中心小学: 南起苏州河，北至天目东路，东起河南北路，西至浙江北路

根据这份数据获取地理信息组成的封闭区域如下：

![](B3D5F6A3-B393-4531-9CA7-38325EB0B2AA.png)

如上图所示，上海市静安区闸北第一中心小学的片区是一个很小的片区，其中还涉及到七浦路服饰批发市场，作为生源招生的住宅在这个片区里并不多。我们分析一下这个片区的组成

分区分析一下整个片区

+ 拆迁区域 - 天目路到海宁路中间这块地段，除了静安区闸北第一中心小学和几幢房子外，已经全部动迁，所以这一片都没有房子出售
+ 七浦路服饰批发市场 - 塘沽路到天潼路中间这块地段，是七浦路服饰批发市场，有些脏乱差，而且小偷很多，一直对这块印象很差，对孩子成长不利
+ 天目路以北的地段，是隶属于之前铁路系统的，有铁路博物馆和铁路系统的福利房
+ 天潼路以南的地段，是这个片区里最好的，富人区，向往～～～

2. 圣和圣广场丽晶馆是商住楼，不能落户

圣和圣广场(公寓)是酒店式公寓，70年产权，但是我不清楚能不能进学区，按照闸北第一中心小学的招生情况看，生源不足，有些租户都可以进，所以酒店式公寓可能没有问题，我暂时把它作为备选。

## 房价

整理一下这些楼盘的均价

```python
t = df[(df['name'].str.contains('天东小区')) | (df['name'].str.contains('河滨豪园')) | ((df['name'].str.contains('圣和圣'))&(df['name'].str.contains('公寓'))) | (df['name'].str.contains('圣和静安公馆'))]
t.groupby(by='name').agg({'price_per_m2':'mean', 'price': 'mean'})
```

从总价格的购买力来说，只有天东小区和圣和圣广场(公寓)符合

## 税费

从税费来看天东小区的税非常便宜，这个小区原来是铁路系统的福利房，根据链家上介绍：

> 此房是铁路局分的售后公房，目前满五唯一，没有个税和增值税，下家首套1.5%,二套3%。


## 居住环境

按照他们的环境做下排序

1. 河滨豪园 - 小区绿化高，配有高档会所，人车分流，3梯2户。 距外滩金融贸易区仅800米,32层高层住宅,三梯两户超前设计,户户面水朝南.物业由港力物业管理(上海)有限公司提供酒店式管理。
2. 天东小区 - 此小区是1999年的小区，是铁路局分配的公房，不仅房屋建筑质量优质，而且居住人员素质也高，居住氛围很好。
3. 圣和静安公馆 - 本小区是2017年的新小区，开发商统一按10000元每平豪壮装修标准交付，适合年轻人购买当作婚房，目前整个小区年轻人居住较多

圣和静安公馆，可能是各人偏好，靠近七浦路地区发展规划应该是向好的，但是周围环境太吵了，不喜欢。

综上：最终买了天东小区，我最终成交的均价在6.4万/平米，这个小区在6.5万/平米左右出房很快。这里展示的数据是2019.12.5的数据，这个90平的房，单价超过7万，截止今天2020.2.17也已经卖了，当前一套房均价也要在7.2万/平米。
